# Encoding

## Label Encoding


```python
# Burada sadece iki tane değer alabilen özellikleri 0 ve 1 e dönüştürdük
col =['class', 'bruises',  'gill-spacing', 'gill-size', 'stalk-shape']
for i in col:
    data[i] = data[i].astype('category')
    data[i] = data[i].cat.codes
```

```python
data.head()
```

```python
data.info()
```

## One Hot encoding

```python
print("Columns having categorical datatype : ")
category_col = list(data.select_dtypes(include = ['object']).columns)
print(category_col)
```

```python
df = pd.get_dummies(data, drop_first =True)
```

# Split Training and Test Data Set


```python
X = df.drop("class",axis=1)
y = df["class"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Models Traininig

## Logistic Regression

```python
m1 ="Logistic Regression"
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
train_pred_lr=lr.predict(X_train)
test_pred_lr =lr.predict(X_test)
```

```python


train_acc_lr =accuracy_score(y_train, train_pred_lr)
test_acc_lr = accuracy_score(y_test, test_pred_lr)

print(f"Train Set Accuracy: {train_acc_lr}")
print(f"Test Set Accuracy: {test_acc_lr}")

# Creating Confusion Matrix
print("\n\nTest Set Confusion Matrix\n")
print(confusion_matrix(y_train,train_pred_lr))

print("\n\nTest Set Confusion Matrix\n")
print(confusion_matrix(y_test,test_pred_lr))
```

```python
print(f"Train Set Report:\n\n {classification_report(y_train,train_pred_lr)}")
print(f"Test Set Report:\n\n {classification_report(y_test, test_pred_lr)}")
```

## Decision Tree Classifier


```python
m2 ="Decision Tree Classifier"
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
train_pred_dtc= dtc.predict(X_train)
test_pred_dtc = dtc.predict(X_test)
```

```python
train_acc_dtc = accuracy_score(y_train, train_pred_dtc)
test_acc_dtc = accuracy_score(y_test, test_pred_dtc)

print(f"Train Set Accuracy: {train_acc_dtc}")
print(f"Test Set Accuracy: {test_acc_dtc}")

# Creating Confusion Matrix
print("\n\nTrain Set Confusion Matrix\n")
print(confusion_matrix(y_train, train_pred_dtc))

print("\n\nTest Set Confusion Matrix\n")
print(confusion_matrix(y_test, test_pred_dtc))
```

```python
print(f"Train Set Report:\n\n {classification_report(y_train, train_pred_dtc)}")
print(f"Test Set Report:\n\n {classification_report(y_test, test_pred_dtc)}")
```

## Random Forest

```python
m3= "Random Forest"
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
train_pred_rf = rf.predict(X_train)
test_pred_rf = rf.predict(X_test)

train_acc_rf = accuracy_score(y_train, train_pred_rf)
test_acc_rf = accuracy_score(y_test, test_pred_rf)


print(f"Train Set Accuracy Score: {train_acc_rf}")
print(f"Test Set Accuracy Score: {test_acc_rf}")

# Creating Confusion Matrix
print("\n\nTrain Set Confusion Matrix\n")
print(confusion_matrix(y_train, train_pred_rf))

print("\n\nTest Set Confusion Matrix\n")
print(confusion_matrix(y_test, test_pred_rf))
```

```python
print(f"Train Set Report:\n\n {classification_report(y_train, train_pred_rf)}")
print(f"Test Set Report:\n\n {classification_report(y_test, test_pred_rf)}")
```

## Support Vector Classification

```python
m4= "Support Vector Classification"
svc = SVC()
svc.fit(X_train, y_train)
```

```python
train_pred_svc = svc.predict(X_train)
test_pred_svc = svc.predict(X_test)

train_acc_svc =accuracy_score(y_train, train_pred_svc)
test_acc_svc= accuracy_score(y_test, test_pred_svc)

print(f"Train Set Accuracy Score: {train_acc_svc}")
print(f"Test Set Accuracy Score: {test_acc_svc}")

# Creating Confusion Matrix
print("\n\nTrain Set Confusion Matrix\n")
print(confusion_matrix(y_train, train_pred_svc))

print("\n\nTest Set Confusion Matrix\n")
print(confusion_matrix(y_test, test_pred_svc))
```

```python
print(f"Train Set Report:\n\n {classification_report(y_train, train_pred_svc)}")
print(f"Test Set Report:\n\n {classification_report(y_test, test_pred_svc)}")
```

## XGB Classifier

```python
m5 = "XGB Classifier"
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
train_pred_xgb = xgb.predict(X_train)
test_pred_xgb = xgb.predict(X_test)

train_acc_xgb = accuracy_score(y_train, train_pred_xgb)
test_acc_xgb = accuracy_score(y_test, test_pred_xgb)

print(f"Train Set Accuracy Score: {train_acc_xgb}")
print(f"Test Set Accuracy Score: {test_acc_xgb}")

# Creating Confusion Matrix
print("\n\nTrain Set Confusion Matrix\n")
print(confusion_matrix(y_train, train_pred_xgb))

print("\n\nTest Set Confusion Matrix\n")
print(confusion_matrix(y_test, test_pred_xgb))

```
